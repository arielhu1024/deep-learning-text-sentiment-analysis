# 04 Model Comparison

This notebook compares the LSTM and Transformer models on SST-5 sentiment classification.

The primary metrics are:

- Accuracy
- Macro F1-score
- Weighted F1-score


```
import pandas as pd
import matplotlib.pyplot as plt

comparison = pd.DataFrame({
    "Model": ["Bidirectional LSTM", "DistilBERT Transformer"],
    "Accuracy": [0.3557, 0.5285],
    "Macro_F1": [0.3259, 0.4850],
    "Weighted_F1": [0.3440, 0.5107]
})

comparison
```

## Performance Visualization

```
ax = comparison.set_index("Model")[["Accuracy", "Macro_F1"]].plot(
    kind="bar",
    figsize=(7, 4),
    rot=0
)

ax.set_title("LSTM vs Transformer Performance on SST-5")
ax.set_ylabel("Score")
ax.set_ylim(0, 0.65)
plt.tight_layout()
plt.show()
```

## Interpretation

The Transformer-based model outperformed the LSTM model across all main evaluation metrics. This suggests that pretrained contextual language representations provide a clear advantage for fine-grained sentiment classification, where subtle semantic differences across five sentiment categories are difficult to capture with a sequence model trained from scratch.
